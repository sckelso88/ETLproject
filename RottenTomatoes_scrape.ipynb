```python
from bs4 import BeautifulSoup
import requests
import pymongo
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.imdb_db
collection = db.items
```

```python
url = "https://www.rottentomatoes.com/top/bestofrt/"
html = requests.get(url).text
```

```python
soup = BeautifulSoup(html, "html.parser")
print(soup.prettify())
```

```python
datas = soup.find_all("table", "table")
```

```python
for data in datas:
    for x in range(100):
        title=data.find_all("a", class_ = "unstyled articleLink")[x].text
        rating=data.find_all("span", class_="tMeterScore")[x].text
        vote=data.find_all("td", class_="right hidden-xs")[x].text
        print(title, rating, vote)
        RT_data = {"Title": title, "Rating": rating, "vote": vote}
        collection.insert_one(RT_data)
```
